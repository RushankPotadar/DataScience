```python
sc_minmax = MinMaxScaler()
sc_minmax
```

```python
## applying Min-Max Scaling to the numerical columns
df[num_features] = sc_minmax.fit_transform(df[num_features])
df[num_features]
```

## 3. Feature Engineering

```python
import numpy as np

df['log_hours_per_week'] = np.log(df['hours_per_week'] + 1) 
```

```python
df['log_hours_per_week']
```

```python
skewed_col = 'capital_gain'  
skewed_col
```

```python
# Calculate IQR for each numerical column
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

# Define outliers (1 = not an outlier, 0 = outlier)
outliers = ((df[num_cols] >= (Q1 - 1.5 * IQR)) & (df[num_cols] <= (Q3 + 1.5 * IQR))).all(axis=1)

df_cleaned = df[outliers]
```

```python
df_cleaned = df[outliers == 1]
```

```python
df_cleaned
```

```python
from sklearn.ensemble import IsolationForest
isolation_forest = IsolationForest(contamination=0.05, random_state=42)
```

```python
isolation_forest
```

```python
outliers = isolation_forest.fit_predict(df[num_cols]) 
outliers
```

```python
is_inlier = outliers == 1
```
