### Standardization of Numerical columns

```python
df_num
```

```python
# Creating an empty dataframe to store the standardize data
scaled_df=pd.DataFrame(columns=['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value'])
```

```python
scaled_df
```

```python
# Calculating the standardize data (Z_score)
## import scipy.stats as stats
## z_score=stats.zscore(df_num) (for auto calculation using package)


## Manual calculation using formula
scaled_df['Volume']=(df_num['Volume']-df_num['Volume'].mean())/df_num['Volume'].std()

scaled_df['Avg Price']=(df_num['Avg Price']-df_num['Avg Price'].mean())/df_num['Avg Price'].std()

scaled_df['Total Sales Value']=(df_num['Total Sales Value']-df_num['Total Sales Value'].mean())/df_num['Total Sales Value'].std()

scaled_df['Discount Rate (%)']=(df_num['Discount Rate (%)']-df_num['Discount Rate (%)'].mean())/df_num['Discount Rate (%)'].std()

scaled_df['Discount Amount']=(df_num['Discount Amount']-df_num['Discount Amount'].mean())/df_num['Discount Amount'].std()

scaled_df['Net Sales Value']=(df_num['Net Sales Value']-df_num['Net Sales Value'].mean())/df_num['Net Sales Value'].std()
```

```python
scaled_df
```

```python
scaled_column_mean=scaled_df.mean()
scaled_column_mean
```

```python
scaled_column_std=scaled_df.std()
scaled_column_std
```

### Data Transfermation i:e converting the catagorical data into dummy variables 

```python
df_cat
```

```python
# Importing packages required for converting
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
```

```python
# Using get_dummies function of pandas package
df_pandas = pd.get_dummies(df_cat, columns=['Date', 'Day'], drop_first=True)
df_pandas
```

```python
df_pandas = pd.get_dummies(df_cat, columns=['SKU'], drop_first=True)
df_pandas
```

```python
one_hot = make_pipeline(OneHotEncoder())
ord_enc = make_pipeline(OrdinalEncoder())
```

```python
## Using one hot encoder 
transform=ColumnTransformer([('one_hot',one_hot,['BU'])])
df=pd.DataFrame(transform.fit_transform(df_cat),columns=transform.get_feature_names_out())
df.head()
```

```python
transform=ColumnTransformer([('one_hot',one_hot,['City'])])
df=pd.DataFrame(transform.fit_transform(df_cat),columns=transform.get_feature_names_out())
df.head()
```

```python
## Using ordinal encoder
transform=ColumnTransformer([('ord_enc',ord_enc,['Brand'])])
df=pd.DataFrame(transform.fit_transform(df_cat),columns=transform.get_feature_names_out())
df.head()
```

```python
transform=ColumnTransformer([('ord_enc',ord_enc,['Model'])])
df=pd.DataFrame(transform.fit_transform(df_cat),columns=transform.get_feature_names_out())
df.head()
```
