### Boxplot

```python
## Creating boxplot to identify outliers and treating it.
df.boxplot()
```

```python
## Defining a function to find IQR, Extreme values, and then to replace it with outliers.
def outlier_removal(df,column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (1.5*IQR)
    upper_bound = Q3 + (1.5*IQR)
    df[column] = df[column].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)
```

```python
outlier_removal(df,'LB')
outlier_removal(df,'ASTV')
outlier_removal(df,'MSTV')
outlier_removal(df,'ALTV')
outlier_removal(df,'MLTV')
outlier_removal(df,'Width')
outlier_removal(df,'Tendency')
outlier_removal(df,'NSP')
```
